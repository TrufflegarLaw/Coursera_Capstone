### Find the nearby venues in KL districts and Tokyo districts within the radius of 5km

```python
radius = 5000
LIMIT = 100



def nearby(df):
    
    venues = []
    
    for name, latitude, longitude in zip(df['District'],df['Latitude'],df['Longitude']):

        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID,CLIENT_SECRET,VERSION,latitude,longitude,radius,LIMIT)

        results = requests.get(url).json()["response"]['groups'][0]['items']

        for v in results:
            venues.append((
                name, 
                latitude, 
                longitude, 
                v['venue']['name'], 
                v['venue']['location']['lat'], 
                v['venue']['location']['lng'],  
                v['venue']['categories'][0]['name']))
    
    return venues
```

### Find the top 10 most common place near the districts

```python
def final_df(df):
    temp_data,header = [],[]
    
    for i in range(0,len(df.iloc[:,0])):
        temp_data.append([])
        for j in range(0,len(df.columns.values)):
            temp_data[i].append(df.iloc[i][j])
            
    header = df.columns.values
    header = header[1:]
    
    result=[]
    for i in range(0,len(temp_data)):
        score = np.array(temp_data[i][1:]).reshape(-1,1)
        temp_header = np.array(header).reshape(-1,1)

        temp_stack = np.column_stack((temp_header,score))
        temp_df = pd.DataFrame(temp_stack,columns=['header','score'])
        temp_df = temp_df.sort_values(['score'],ascending=0).reset_index(drop=True)[0:10]
        temp_df = temp_df.iloc[:,0]
        result.append(temp_df)

    nbh = np.array(df['District']).reshape(-1,1)
    final_data = np.column_stack((nbh,result))
    ranked_df = pd.DataFrame(final_data, columns = ['District','1st','2nd','3rd','4th','5th',
                                                  '6th','7th','8th','9th','10th'])
    return(ranked_df)
```

## CLUSTERING AND SEGMENTATION

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
clustering_df = avg.drop(columns = ['District'])
```

```python
# Elbow Method
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, random_state=111, init = 'k-means++')
    kmeans.fit(clustering_df)
    wcss.append(kmeans.inertia_)
    print('No =',i,', SSE =', kmeans.inertia_)
    
    
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(14,8))
plt.plot(range(1,11),wcss, color = 'blue')
plt.xlabel('No of Clusters')
plt.ylabel('Sum of Square (error)')
plt.xticks(range(1,11))
plt.show()
```

```python
# from the figure above, we found out that the  best kmeans model is when no. of clusters = 3
# Lets recreate  the mode

kmeans = KMeans(n_clusters = 3, random_state = 111, init = 'k-means++')
kmeans.fit(clustering_df)
y = kmeans.labels_
y
```

```python
def cluster_hist (df, labels, color ):
    cluster_df = df
    cluster_df = cluster_df.drop(columns=['District','Latitude','Longitude','Cluster'])
    cluster_df
    
    category = cluster_df.values.tolist()
    temp2 = []
    for i in range(0,len(df.iloc[:,0])):
        for j in range(0,10):
            temp1 = category[i][j]
            temp2.append(temp1)

    unique = list(dict.fromkeys(temp2))
    bins = len(unique)

    plt.figure(figsize=(14,8))
    plt.hist(temp2, bins = bins, color = color, histtype = 'bar',edgecolor = 'white' )
    plt.xticks(rotation = 90)
    plt.yticks(range(0,len(df.iloc[:,0])+1))
    plt.title(labels)
    plt.show()
```
