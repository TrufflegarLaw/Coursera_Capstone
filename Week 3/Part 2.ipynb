```python
Lat_list = np.array(round(df1['Latitude'],6)).reshape(1,-1)
Lng_list = np.array(round(df1['Longitude'],6)).reshape(1,-1)
data = np.column_stack((pc[0],Lat_list[0],Lng_list[0]))
df2 = pd.DataFrame(data,columns=['Postcode','Latitude','Longitude'])
df2
```
